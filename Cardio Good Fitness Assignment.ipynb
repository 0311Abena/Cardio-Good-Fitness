Multivariate Data Analysis
- Analysis of interaction between features, in the dataset    

```python
#using only numerical categories
sns.pairplot(data = cardio_df)
```

```python
# Select only numerical columns
num_cols = cardio_df.select_dtypes(include=['int64']).columns

# Calculate correlation matrix
corr_matrix = cardio_df[num_cols].corr()

# Visualize correlation matrix using heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
```

```python
cardio_df.corr
```

The heatmap shows the paired correlation between every column in the dataset. The correlation coefficient ranges from -1 to 1, where -1 implies a perfect negative correlation, 0 implies no correlation, and 1 implies a perfect positive correlation.

Based on the heatmap, we can see that there is a moderate positive correlation between the following pairs of columns:

Usage and Fitness (correlation coefficient of 0.67)
Fitness and Miles (correlation coefficient of 0.79)

We notice a weak positive correlation between Income and Miles (correlation coefficient of 0.51).

In conclusion, there is no strong correlation between any of the columns.
